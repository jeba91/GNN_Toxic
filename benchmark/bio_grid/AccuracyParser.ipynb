```python
from os import listdir
from os.path import isfile, join
import os
mypath = os.getcwd()
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
onlyfiles = sorted(onlyfiles)
for f in onlyfiles:
    print(f)
```

```python
accs = []
for file_txt in onlyfiles:
#     print(file_txt)
    if file_txt[-3:] == "txt":
        file = open(file_txt, mode = 'r', encoding = 'utf-8-sig')
        lines = file.readlines()
        file.close()
        printed=False
        train = 0
        for l in lines:
            if 'main_TUs' in str(l):
                name = str(l).strip().split(" ")[3]
            if "TRAIN ACCURACY averaged" in str(l) and "nan" not in str(l):
                train = str(l).strip().split(" ")[3]
        if train != 0:
            accs.append([name, train, file_txt])
#             printed = True
#         if not printed:
#             print(file_txt)
```

```python
import pandas as pd
ac = pd.DataFrame(accs)
```

```python
ac = ac.sort_values(by=[0])
for a in list(set(ac[0].values)):
    print(a, max(ac[1].loc[ac[0] == a]))
    print(str(ac[2].loc[ac[1] == (max(ac[1].loc[ac[0] == a]))]))
```
